# Analysis of Andaman And Nicobar Islands

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Andaman&nicobar.xlsx')
```

```python
data.head()
```

```python
data.tail()
```

# January Analysis 

```python
import seaborn as sns
sns.set()
plt.plot(data['YEAR'],data['JAN'])
plt.scatter(data['YEAR'],data['JAN'])
plt.xlabel('Year')
plt.ylabel('MM')
plt.title('January Analysis')
plt.show()
```
