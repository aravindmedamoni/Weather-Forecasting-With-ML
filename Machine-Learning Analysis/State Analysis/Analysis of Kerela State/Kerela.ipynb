```python
Ind = pd.read_csv('rainfall in india 1901-2015.csv')
```

```python
Ind.isnull().sum()
```

```python
Ind['JAN'].fillna((Ind['JAN'].mean()), inplace=True)
Ind['FEB'].fillna((Ind['FEB'].mean()), inplace=True)
Ind['MAR'].fillna((Ind['MAR'].mean()), inplace=True)
Ind['APR'].fillna((Ind['APR'].mean()), inplace=True)
Ind['MAY'].fillna((Ind['MAY'].mean()), inplace=True)
Ind['JUN'].fillna((Ind['JUN'].mean()), inplace=True)
Ind['JUL'].fillna((Ind['JUL'].mean()), inplace=True)
Ind['AUG'].fillna((Ind['AUG'].mean()), inplace=True)
Ind['SEP'].fillna((Ind['SEP'].mean()), inplace=True)
Ind['OCT'].fillna((Ind['OCT'].mean()), inplace=True)
Ind['NOV'].fillna((Ind['NOV'].mean()), inplace=True)
Ind['DEC'].fillna((Ind['DEC'].mean()), inplace=True)
Ind['Jan-Feb'].fillna((Ind['Jan-Feb'].mean()), inplace=True)
Ind['Mar-May'].fillna((Ind['Mar-May'].mean()), inplace=True)
Ind['Jun-Sep'].fillna((Ind['Jun-Sep'].mean()), inplace=True)
Ind['Oct-Dec'].fillna((Ind['Oct-Dec'].mean()), inplace=True)
Ind['ANNUAL'].fillna((Ind['ANNUAL'].mean()), inplace=True)
```

```python
Ind.isnull().sum()
```

```python
I=Ind.groupby("YEAR").sum()['ANNUAL'].plot(figsize=(20,10));
I.set_xlabel('Year')
I.set_ylabel('Annual Rainfall in mm')
I.set_title('Average annual rainfall in India')
```

```python
S = Dist.sort_values('ANNUAL',ascending=True)
S.head()
```

```python
fig = plt.figure(figsize=(15, 10))
ax = plt.xticks(rotation=90)
ax = sns.barplot(x='STATE_UT_NAME', y='ANNUAL', data=S)
plt.ylabel('State/UT')
plt.xlabel('Annual rainfall in mm')
plt.title('Annual rainfall in different states of India')
```

```python
a=Ind[['YEAR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("YEAR").sum().plot(figsize=(20,10));
a.set_xlabel('Year')
a.set_ylabel('Annual Rainfall in mm')
a.set_title('Monthly rainfall over last 115 years')
```

```python
b=Ind[['YEAR','Jan-Feb', 'Mar-May','Jun-Sep', 'Oct-Dec']].groupby("YEAR").sum().plot(figsize=(20,15));
b.set_xlabel('Year')
b.set_ylabel('Annual Rainfall in mm')
b.set_title('Rainfall during different times of Year in India')
```
