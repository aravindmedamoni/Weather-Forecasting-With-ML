# Complete Analysis of RainFall Data

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('district wise rainfall normal.csv') 
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
KerelaDist = dataset[dataset['STATE_UT_NAME'] == 'KERALA']
KerelaDist
```

```python
completeIndia = pd.read_csv('rainfall in india 1901-2015.csv')
```

```python
completeIndia.head()
```

```python
completeIndia.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Kerala=completeIndia[completeIndia['SUBDIVISION']=='KERALA']
Kerala
```

```python
k = Kerala.plot('YEAR', 'ANNUAL')
k.set_xlabel('Year')
k.set_ylabel('Annual Rainfall in mm')
k.set_title('Average annual rainfall')
```
